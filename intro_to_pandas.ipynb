```python
#we sometimes often have missing data in pur dataset in realworld not in this example in that case

car_sales_missing = pd.read_csv("data/9.2 car-sales-missing-data.csv")
car_sales_missing
#NaN Missing there are no values in that place
```

```python
car_sales_missing["Odometer"].mean()
```

```python
#So to fill in values
#fillna() can fill in NA values with non-NA data in acouple of ways
car_sales_missing["Odometer"].fillna(car_sales_missing["Odometer"].mean())
```

```python
car_sales_missing
```

```python
#inplace replaces all the reassigning thing with this parameter
car_sales_missing["Odometer"].fillna(car_sales_missing["Odometer"].mean(),inplace=True)
```

```python
car_sales_missing
```

```python
#if we want to get rid of the NAN values
car_sales_missing.dropna(inplace=True)
```

```python
car_sales_missing
```

```python
#To regain the old csv we have to reimport it to get old columns back
```

```python
car_sales_missing_dropped = car_sales_missing.dropna()
car_sales_missing_dropped
```
